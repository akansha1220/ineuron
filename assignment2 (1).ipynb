```python
# TASK - 1
# question 1.1

"""Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()"""

def myreduce(func,collection):
    result = collection[0]
    for i in range(1,len(collection)):
        result = func(result,collection[i])
    return result

def func(a,b):
    return a+b;

l = [1,2,3,4,5]
print("this is myreduce function: ",myreduce(func,l))
        
```

```python
# question 1.2
"""Write a Python program to implement your own myfilter() function which works exactly like
Python's built-in function filter()"""

def myfilter(func1,collection):
    result=[]
    for j in collection:
        if(func1(j)):
            result.append(j)
    return result

def func1(a):
    if(a>0):
        return True
    else:
        return False
    
ll = [1,2,3,-1,-4,-6,13,-24]
print("this is my filter function: ",myfilter(func1,ll))   
    
```

```python
#task 1 problem3
# longestWord()

from functools import reduce
def LongestWord(L):
    M=0
    
    for i in L:
        if len(i)>M:
            M=len(i)
            hig=i
    return hig

L=[]

p=int(input("Enter The List Length : "))
print("Enters The Words ")
for i in range(p):
    s=input(" ")
    L.append(s)
    
print("Longest word in the list : ",LongestWord(L))   
print("By Using Reduce Function ",reduce(lambda x,y:x if len(x)>len(y) else y,L))

```
